```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load Data
try:
    df = pd.read_csv('people.csv', encoding='cp949')
except UnicodeDecodeError:
    df = pd.read_csv('people.csv', encoding='utf-8')

# 2. Process Nationwide Data
nation_row = df.iloc[0]
pop_cols = [c for c in df.columns if '_총인구수' in c]

dates = []
pops = []

for col in pop_cols:
    date_str = col.split('_')[0] # e.g., 2021년01월
    pop_val = int(str(nation_row[col]).replace(',', ''))

    dates.append(date_str)
    pops.append(pop_val)

data = pd.DataFrame({'Date_Str': dates, 'Population': pops})

# Convert to datetime
data['Date'] = pd.to_datetime(data['Date_Str'], format='%Y년%m월')
data['Year'] = data['Date'].dt.year
data['Month'] = data['Date'].dt.month

# Calculate Monthly Net Change
data['Net_Change'] = data['Population'].diff()

# 3. Visualization: Facet Grid of Bar Charts
years = data['Year'].unique()
n_years = len(years)

fig, axes = plt.subplots(n_years, 1, figsize=(12, 4 * n_years))
# Handle case if only 1 year (though we know there are 5)
if n_years == 1:
    axes = [axes]

for i, year in enumerate(years):
    year_data = data[data['Year'] == year]
    ax = axes[i]

    # Create bar plot
    # We remove Jan if it is NaN (which is the case for the very first month of dataset)
    plot_data = year_data.dropna(subset=['Net_Change'])

    colors = ['red' if x < 0 else 'blue' for x in plot_data['Net_Change']]
    bars = ax.bar(plot_data['Month'], plot_data['Net_Change'], color=colors, alpha=0.7)

    ax.set_title(f'{year} Monthly Population Change', fontsize=14, fontweight='bold')
    ax.set_ylabel('Net Change')
    ax.set_xlabel('Month')
    ax.set_xticks(range(1, 13))
    ax.grid(axis='y', linestyle='--', alpha=0.5)

    # Add value labels on top of bars
    for bar in bars:
        height = bar.get_height()
        xy_pos = (bar.get_x() + bar.get_width() / 2, height)
        text_pos = (0, -12) if height < 0 else (0, 5) # Text below bar if negative

        ax.annotate(f'{int(height):,}',
                    xy=xy_pos,
                    xytext=text_pos,
                    textcoords="offset points",
                    ha='center', va='center', fontsize=9)

plt.tight_layout()
plt.savefig('yearly_monthly_fluctuations.png')
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load and Prepare Data
# Re-loading as the environment is stateless between turns for variables not saved.
try:
    df = pd.read_csv('people.csv', encoding='cp949')
except UnicodeDecodeError:
    df = pd.read_csv('people.csv', encoding='utf-8')

# Extract Nationwide data (first row)
nation_row = df.iloc[0]
pop_cols = [c for c in df.columns if '_총인구수' in c]
dates_str = [c.split('_')[0] for c in pop_cols]
pop_values = nation_row[pop_cols].values
pop_values = [int(str(x).replace(',', '')) for x in pop_values]

data = pd.DataFrame({
    'Date_Str': dates_str,
    'Population': pop_values
})

# Convert Date
data['Date'] = pd.to_datetime(data['Date_Str'], format='%Y년%m월')
data['Year'] = data['Date'].dt.year
data['Month'] = data['Date'].dt.month
data['Timestamp'] = data['Date'].map(pd.Timestamp.toordinal) # continuous time variable if needed, but Year+Month dummies works too.

# 2. Feature Engineering: One-Hot Encoding for Month
# User asked for One-Hot Encoding.
# We will encode 'Month'. 'Year' will remain numeric to capture the trend.
data_encoded = pd.get_dummies(data, columns=['Month'], prefix='Month', drop_first=True)
# drop_first=True to avoid multicollinearity (dummy variable trap), though LinearRegression in sklearn handles it okay, it's good practice.
# Actually, let's keep all or drop first. drop_first=True is safer.

# Prepare X and y
# Features: Year, and the Month dummies
feature_cols = [c for c in data_encoded.columns if c.startswith('Month_') or c == 'Year']
X = data_encoded[feature_cols]
y = data_encoded['Population']

# 3. Train/Test Split
# Since this is time series data, we should not shuffle. We split by time.
# Let's use the last 12 months for testing to see how well it predicts recent trends.
split_index = len(data) - 12
X_train = X.iloc[:split_index]
y_train = y.iloc[:split_index]
X_test = X.iloc[split_index:]
y_test = y.iloc[split_index:]

# 4. Model Training
model = LinearRegression()
model.fit(X_train, y_train)

# 5. Evaluation
y_pred_test = model.predict(X_test)

mae = mean_absolute_error(y_test, y_pred_test)
mse = mean_squared_error(y_test, y_pred_test)
rmse = np.sqrt(mse)

print(f"Evaluation Metrics:")
print(f"MAE: {mae:,.2f}")
print(f"MSE: {mse:,.2f}")
print(f"RMSE: {rmse:,.2f}")

# 6. Future Prediction (up to 2050)
last_date = data['Date'].max()
future_dates = pd.date_range(start=last_date + pd.DateOffset(months=1), end='2050-12-01', freq='MS')

future_data = pd.DataFrame({
    'Date': future_dates
})
future_data['Year'] = future_data['Date'].dt.year
future_data['Month'] = future_data['Date'].dt.month

# Apply same One-Hot Encoding
future_encoded = pd.get_dummies(future_data, columns=['Month'], prefix='Month', drop_first=True)

# Ensure all columns from training exist in future data (e.g. if some months were missing, which is unlikely here)
# Reindex to match X columns, filling 0 for missing columns (if any)
future_X = future_encoded.reindex(columns=feature_cols, fill_value=0)

future_pred = model.predict(future_X)
future_data['Predicted_Population'] = future_pred

# 7. Visualization
plt.figure(figsize=(14, 7))

# Historical Data
plt.plot(data['Date'], data['Population'], label='Historical Data', color='black')

# Test Predictions
test_dates = data['Date'].iloc[split_index:]
plt.plot(test_dates, y_pred_test, label='Test Set Predictions (Linear Regression)', color='orange', linestyle='--')

# Future Predictions
plt.plot(future_data['Date'], future_data['Predicted_Population'], label='Future Prediction (2050)', color='red', linestyle='--')

plt.title('Population Prediction up to 2050 using Linear Regression')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend()
plt.grid(True)
plt.savefig('population_prediction_2050.png')

# Show some future values
print("\nFuture Predictions (Head):")
print(future_data[['Date', 'Predicted_Population']].head())
print("\nFuture Predictions (Tail):")
print(future_data[['Date', 'Predicted_Population']].tail())
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

# 1. Load Data
try:
    df = pd.read_csv('people.csv', encoding='cp949')
except UnicodeDecodeError:
    df = pd.read_csv('people.csv', encoding='utf-8')

# Process
nation_row = df.iloc[0]
pop_cols = [c for c in df.columns if '_총인구수' in c]
dates_str = [c.split('_')[0] for c in pop_cols]
pop_values = [int(str(nation_row[c]).replace(',', '')) for c in pop_cols]

data = pd.DataFrame({'Date': pd.to_datetime(dates_str, format='%Y년%m월'), 'Population': pop_values})
data['Year'] = data['Date'].dt.year
data['Month'] = data['Date'].dt.month
data['Ordinal'] = data['Date'].map(pd.Timestamp.toordinal)

# 2. Prepare Data for Models
# Use Ordinal date for simpler plotting and polynomial fitting (Year+Month dummies is good, but Ordinal is easier for Poly)
# Let's stick to the previous feature set (Year + Month Dummies) for Linear/Ridge/Lasso consistency.
# But for Polynomial, we usually poly transform the continuous variable (Year or Ordinal).
# Let's use 'Ordinal' as the main time feature for all, plus Month dummies for seasonality.

# Feature Engineering
data_encoded = pd.get_dummies(data, columns=['Month'], prefix='Month', drop_first=True)
# We will use 'Ordinal' instead of 'Year' to be more precise with time continuity for Poly regression
feature_cols = ['Ordinal'] + [c for c in data_encoded.columns if c.startswith('Month_')]

X = data_encoded[feature_cols]
y = data_encoded['Population']

# Split (9:1)
split_idx = int(len(data) * 0.9)
X_train, X_test = X.iloc[:split_idx], X.iloc[split_idx:]
y_train, y_test = y.iloc[:split_idx], y.iloc[split_idx:]

# 3. Define Models
models = {
    'Linear Regression': LinearRegression(),
    'Ridge (alpha=1.0)': Ridge(alpha=1.0),
    'Lasso (alpha=1.0)': Lasso(alpha=1.0),
    'Polynomial (Deg 2)': make_pipeline(PolynomialFeatures(degree=2), LinearRegression())
}

# 4. Train & Predict
future_preds = {}
test_preds = {}

# Create Future Data
last_date = data['Date'].max()
future_dates = pd.date_range(start=last_date + pd.DateOffset(months=1), end='2050-12-01', freq='MS')
future_df = pd.DataFrame({'Date': future_dates})
future_df['Ordinal'] = future_df['Date'].map(pd.Timestamp.toordinal)
future_df['Month'] = future_df['Date'].dt.month
future_encoded = pd.get_dummies(future_df, columns=['Month'], prefix='Month', drop_first=True)
future_X = future_encoded.reindex(columns=feature_cols, fill_value=0)

for name, model in models.items():
    # Train on Full Train set (90%)
    model.fit(X_train, y_train)
    test_preds[name] = model.predict(X_test)

    # Train on Full Data for Future Forecast
    model.fit(X, y)
    future_preds[name] = model.predict(future_X)

# 5. Visualization - Detailed Plots
fig, axes = plt.subplots(3, 1, figsize=(12, 18))

# Plot 1: Full Range (2021-2050)
axes[0].plot(data['Date'], data['Population'], 'k-', label='Historical Data', linewidth=2)
for name, preds in future_preds.items():
    axes[0].plot(future_df['Date'], preds, label=name, linestyle='--')
axes[0].set_title('1. Long-term Prediction (2021-2050)', fontsize=14, fontweight='bold')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# Plot 2: Test Set Zoom (Recent 6 Months)
axes[1].plot(data['Date'].iloc[split_idx:], y_test, 'k-', label='Actual Data', marker='o', linewidth=2)
for name, preds in test_preds.items():
    axes[1].plot(data['Date'].iloc[split_idx:], preds, label=name, marker='x', linestyle='--')
axes[1].set_title('2. Model Accuracy Check (Recent 6 Months)', fontsize=14, fontweight='bold')
axes[1].legend()
axes[1].grid(True, alpha=0.3)
# Calculate limits to make it distinct
min_y = min(y_test.min(), min([min(p) for p in test_preds.values()]))
max_y = max(y_test.max(), max([max(p) for p in test_preds.values()]))
axes[1].set_ylim(min_y - 10000, max_y + 10000)


# Plot 3: Future Zoom (2045-2050) - Divergence
axes[2].set_title('3. Long-term Divergence (2045-2050 Zoom)', fontsize=14, fontweight='bold')
for name, preds in future_preds.items():
    # Filter for 2045 onwards
    mask = future_df['Date'].dt.year >= 2045
    axes[2].plot(future_df.loc[mask, 'Date'], preds[mask], label=name, linestyle='-', linewidth=2)
axes[2].legend()
axes[2].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('linear_models_detailed_analysis.png')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.metrics import mean_absolute_error, mean_squared_error

# 1. Load Data
try:
    df = pd.read_csv('people.csv', encoding='cp949')
except UnicodeDecodeError:
    df = pd.read_csv('people.csv', encoding='utf-8')

# Process Nationwide Data
nation_row = df.iloc[0]
pop_cols = [c for c in df.columns if '_총인구수' in c]
dates_str = [c.split('_')[0] for c in pop_cols]
pop_values = [int(str(nation_row[c]).replace(',', '')) for c in pop_cols]

data = pd.DataFrame({'Date': pd.to_datetime(dates_str, format='%Y년%m월'), 'Population': pop_values})
data['Year'] = data['Date'].dt.year
data['Month'] = data['Date'].dt.month
data['Ordinal'] = data['Date'].map(pd.Timestamp.toordinal)
data = data.sort_values('Date')

# 2. Split Data (Train: 2021-2024, Test: 2025)
train = data[data['Year'] <= 2024].copy()
test = data[data['Year'] == 2025].copy()

print(f"Train samples: {len(train)}")
print(f"Test samples: {len(test)}")

# 3. Train Models
# Feature Engineering
# We need to make sure dummies are consistent.
# We'll create dummies for the whole dataset first or handle it carefully.
# Easier to fit OneHotEncoder or use pd.get_dummies on combined, then split.
data_encoded = pd.get_dummies(data, columns=['Month'], prefix='Month', drop_first=True)
feature_cols = ['Ordinal'] + [c for c in data_encoded.columns if c.startswith('Month_')]

X = data_encoded[feature_cols]
y = data_encoded['Population']

X_train = X[data['Year'] <= 2024]
y_train = y[data['Year'] <= 2024]
X_test = X[data['Year'] == 2025]
y_test = y[data['Year'] == 2025]

# Models
models = {
    'Linear Regression': LinearRegression(),
    'Ridge': Ridge(alpha=1.0),
    'Lasso': Lasso(alpha=1.0),
    'Polynomial (Deg 2)': make_pipeline(PolynomialFeatures(degree=2), LinearRegression())
}

metrics = {}
for name, model in models.items():
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    mae = mean_absolute_error(y_test, preds)
    rmse = np.sqrt(mean_squared_error(y_test, preds))
    metrics[name] = {'MAE': mae, 'RMSE': rmse}

print("Evaluation on 2025 Test Data:")
print(pd.DataFrame(metrics).T)

# 4. Future Prediction (up to 2070)
future_dates = pd.date_range(start='2021-01-01', end='2070-12-01', freq='MS')
future_df = pd.DataFrame({'Date': future_dates})
future_df['Year'] = future_df['Date'].dt.year
future_df['Month'] = future_df['Date'].dt.month
future_df['Ordinal'] = future_df['Date'].map(pd.Timestamp.toordinal)

future_encoded = pd.get_dummies(future_df, columns=['Month'], prefix='Month', drop_first=True)
future_X = future_encoded.reindex(columns=feature_cols, fill_value=0)

future_preds = {}
for name, model in models.items():
    future_preds[name] = model.predict(future_X)

# 5. Load Official Projection
try:
    df_proj = pd.read_csv('미래 인구 전망(총 인구).csv')
except:
    df_proj = pd.read_csv('미래 인구 전망(총 인구).csv', encoding='cp949')

proj_row = df_proj[df_proj['구분'] == '인구'].iloc[0]
proj_years = [c for c in df_proj.columns if c.isdigit()]
proj_values_clean = []
for x in proj_row[proj_years].values:
    try:
        proj_values_clean.append(int(float(str(x).replace(',', ''))))
    except:
        proj_values_clean.append(np.nan)

data_proj = pd.DataFrame({'Year': [int(y) for y in proj_years], 'Population': proj_values_clean})
data_proj['Date'] = pd.to_datetime(data_proj['Year'], format='%Y')

# 6. Visualization
plt.figure(figsize=(14, 8))
plt.rc('font', family='NanumBarunGothic')
plt.rcParams['axes.unicode_minus'] = False

# Historical (Train)
plt.plot(train['Date'], train['Population'], 'k-', label='Train Data (2021-2024)', linewidth=2)
# Test (Actual)
plt.plot(test['Date'], test['Population'], 'k--', label='Test Data (Actual 2025)', linewidth=2)

# Models
colors = {'Linear Regression': 'red', 'Ridge': 'magenta', 'Lasso': 'purple', 'Polynomial (Deg 2)': 'green'}
styles = {'Linear Regression': '--', 'Ridge': ':', 'Lasso': '-.', 'Polynomial (Deg 2)': '-.'}

for name, preds in future_preds.items():
    # Plot only from 2025 onwards for predictions? Or full range?
    # Let's plot full range to show fit on train too?
    # Usually strictly prediction part.
    # Let's plot from 2025 to 2070
    mask = future_df['Year'] >= 2025
    plt.plot(future_df.loc[mask, 'Date'], preds[mask], label=f'{name}', color=colors[name], linestyle=styles[name])

# Official
data_proj_2070 = data_proj[(data_proj['Year'] >= 2020) & (data_proj['Year'] <= 2070)]
plt.plot(data_proj_2070['Date'], data_proj_2070['Population'], label='Official Projection', color='orange', marker='o', markersize=6)

# Annotations
val_2070 = {}
for name, preds in future_preds.items():
    val_2070[name] = preds[-1]
val_official_2070 = data_proj[data_proj['Year'] == 2070]['Population'].values[0] if 2070 in data_proj['Year'].values else np.nan

end_date = future_df['Date'].iloc[-1]
def add_label(date, val, text, color, offset=(0,0)):
    plt.annotate(f"{text}: {int(val/10000):,}만", (date, val), xytext=offset, textcoords='offset points', ha='left', fontsize=10, color=color, fontweight='bold', bbox=dict(boxstyle="round,pad=0.3", fc="white", ec=color, alpha=0.8))

add_label(end_date, val_2070['Linear Regression'], "Linear", "red", (10, 0))
add_label(end_date, val_2070['Polynomial (Deg 2)'], "Polynomial", "green", (10, -20))
# Ridge/Lasso near Linear
add_label(end_date, val_2070['Ridge'], "Ridge", "magenta", (10, 20))
add_label(pd.Timestamp('2070-01-01'), val_official_2070, "Official", "orange", (10, 0))

plt.title('Population Prediction (Train: 2021-2024, Test: 2025) vs Official Projection', fontsize=16)
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('train_2024_test_2025_analysis.png')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# 1. Load Data
try:
    df_hist = pd.read_csv('people.csv', encoding='cp949')
except UnicodeDecodeError:
    df_hist = pd.read_csv('people.csv', encoding='utf-8')

# Process
nation_row = df_hist.iloc[0]
pop_cols = [c for c in df_hist.columns if '_총인구수' in c]
dates_str = [c.split('_')[0] for c in pop_cols]
pop_values = [int(str(nation_row[c]).replace(',', '')) for c in pop_cols]

data = pd.DataFrame({'Date': pd.to_datetime(dates_str, format='%Y년%m월'), 'Population': pop_values})
data['Year'] = data['Date'].dt.year
data['Month'] = data['Date'].dt.month
data['Ordinal'] = data['Date'].map(pd.Timestamp.toordinal)
data = data.sort_values('Date')

# 2. Prepare Features (All Data 2021-2025)
X = pd.get_dummies(data, columns=['Month'], prefix='Month', drop_first=True)
feature_cols = ['Ordinal'] + [c for c in X.columns if c.startswith('Month_')]
X = X[feature_cols]
y = data['Population']

# 3. Train Models
models = {
    'Linear Regression': LinearRegression(),
    'Ridge': Ridge(alpha=1.0),
    'Lasso': Lasso(alpha=1.0),
    'Polynomial (Deg 2)': make_pipeline(PolynomialFeatures(degree=2), LinearRegression())
}

results = []

for name, model in models.items():
    # Fit
    model.fit(X, y)
    # Predict (In-sample prediction)
    y_pred = model.predict(X)

    # Calculate Metrics
    mae = mean_absolute_error(y, y_pred)
    mse = mean_squared_error(y, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y, y_pred)

    # MAPE (Mean Absolute Percentage Error)
    mape = np.mean(np.abs((y - y_pred) / y)) * 100

    results.append({
        'Model': name,
        'MAE (명)': mae,
        'RMSE (명)': rmse,
        'MAPE (%)': mape,
        'R² Score': r2
    })

results_df = pd.DataFrame(results).set_index('Model')

# 4. Visualization - Residual Plot (Error over time)
plt.figure(figsize=(14, 6))
plt.rc('font', family='NanumBarunGothic')
plt.rcParams['axes.unicode_minus'] = False

colors = {'Linear Regression': 'red', 'Ridge': 'magenta', 'Lasso': 'purple', 'Polynomial (Deg 2)': 'green'}

for name, model in models.items():
    y_pred = model.predict(X)
    residuals = y - y_pred # Actual - Predicted
    plt.plot(data['Date'], residuals, label=name, marker='o', markersize=4, linestyle='--', color=colors[name], alpha=0.7)

plt.axhline(0, color='black', linewidth=1)
plt.title('Residual Analysis (2021-2025): Actual - Predicted', fontsize=16)
plt.ylabel('Error (Population Count)')
plt.xlabel('Date')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('model_evaluation_metrics.png')

print("Model Evaluation Metrics (2021-2025):")
print(results_df)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# 1. Load Data
try:
    df_hist = pd.read_csv('people.csv', encoding='cp949')
except UnicodeDecodeError:
    df_hist = pd.read_csv('people.csv', encoding='utf-8')

# Process
nation_row = df_hist.iloc[0]
pop_cols = [c for c in df_hist.columns if '_총인구수' in c]
dates_str = [c.split('_')[0] for c in pop_cols]
pop_values = [int(str(nation_row[c]).replace(',', '')) for c in pop_cols]

data = pd.DataFrame({'Date': pd.to_datetime(dates_str, format='%Y년%m월'), 'Population': pop_values})
data['Year'] = data['Date'].dt.year
data['Month'] = data['Date'].dt.month
data['Ordinal'] = data['Date'].map(pd.Timestamp.toordinal)
data = data.sort_values('Date')

# 2. Prepare Features
X = pd.get_dummies(data, columns=['Month'], prefix='Month', drop_first=True)
feature_cols = ['Ordinal'] + [c for c in X.columns if c.startswith('Month_')]
X = X[feature_cols]
y = data['Population']

# 3. Train/Test Split (80% : 20%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)

# 4. Train Models & Evaluate
models = {
    'Linear Regression': LinearRegression(),
    'Ridge': Ridge(alpha=1.0),
    'Lasso': Lasso(alpha=1.0),
    'Polynomial (Deg 2)': make_pipeline(PolynomialFeatures(degree=2), LinearRegression())
}

results = []

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)
    mape = np.mean(np.abs((y_test - y_pred) / y_test)) * 100

    results.append({
        'Model': name,
        'MAE (명)': mae,
        'RMSE (명)': rmse,
        'MAPE (%)': mape,
        'R² Score': r2
    })

results_df = pd.DataFrame(results).set_index('Model')
pd.set_option('display.float_format', '{:.6f}'.format)
print("Evaluation Metrics (Test Set - 20%):")
print(results_df)
```
